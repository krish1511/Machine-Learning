```python
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
x= np.arange(0,10)
y= np.arange(11,21)
```

```python
a = np.arange(40,50)
b = np.arange(60,70)
```

```python
plt.scatter(x,y,c='r')
plt.show()
```

```python
y = x*x
plt.scatter(x,y,c='b')
plt.xlabel('x axis')
plt.ylabel('y axis')
plt.title('Graph in 2D')
```

```python
plt.plot(x,y,'r*',linestyle='dashed',linewidth=2,markersize=15)
```

```python
plt.subplot(2,2,1)
plt.plot(x,y,'r')
plt.subplot(2,2,2)
plt.plot(x,y,'g')
plt.subplot(2,2,3)
plt.plot(x,y,'b')
```

```python
plt.plot([1,2,3,4],[8,5,1,6],'r*')
plt.grid()
plt.show()
```

```python
t = np.arange(0,5,0.2,dtype = float)
plt.plot(t,t**2,'b--',label='^2')
plt.plot(t,t**2.2,'r*',label='^2.2')
plt.plot(t,t**2.5,'g^',label='^2.5')
plt.grid()
plt.legend()
plt.show()
```

```python
x1=[1,2,3,4]
y1=[1,4,9,16]
x2=[1,2,3,4]
y2=[2,4,6,8]
lines=plt.plot(x1,y1,x2,y2)
plt.setp(lines[0],color='r',linewidth=2.0)
plt.setp(lines[1],color='b',linewidth=2.0)
plt.grid()
plt.show()
```

```python
plt.figure(1)
plt.subplot(211)
plt.plot([1,2,3])
plt.subplot(212)
plt.plot([4,5,6])

plt.figure(2)
plt.plot([1,5,9])

plt.figure(1)
plt.subplot(211)
plt.title('title as 1,2,3')
plt.show()
```
